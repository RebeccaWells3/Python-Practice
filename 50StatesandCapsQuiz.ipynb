```python
#capitols of 50 states quiz game 
#use random library to use .choice to randomly select a state

#import random library
import random

#use dictionary 
us_states_capitals = {
    "Alabama": "Montgomery",
    "Alaska": "Juneau",
    "Arizona": "Phoenix",
    "Arkansas": "Little Rock",
    "California": "Sacramento",
    "Colorado": "Denver",
    "Connecticut": "Hartford",
    "Delaware": "Dover",
    "Florida": "Tallahassee",
    "Georgia": "Atlanta",
    "Hawaii": "Honolulu",
    "Idaho": "Boise",
    "Illinois": "Springfield",
    "Indiana": "Indianapolis",
    "Iowa": "Des Moines",
    "Kansas": "Topeka",
    "Kentucky": "Frankfort",
    "Louisiana": "Baton Rouge",
    "Maine": "Augusta",
    "Maryland": "Annapolis",
    "Massachusetts": "Boston",
    "Michigan": "Lansing",
    "Minnesota": "Saint Paul",
    "Mississippi": "Jackson",
    "Missouri": "Jefferson City",
    "Montana": "Helena",
    "Nebraska": "Lincoln",
    "Nevada": "Carson City",
    "New Hampshire": "Concord",
    "New Jersey": "Trenton",
    "New Mexico": "Santa Fe",
    "New York": "Albany",
    "North Carolina": "Raleigh",
    "North Dakota": "Bismarck",
    "Ohio": "Columbus",
    "Oklahoma": "Oklahoma City",
    "Oregon": "Salem",
    "Pennsylvania": "Harrisburg",
    "Rhode Island": "Providence",
    "South Carolina": "Columbia",
    "South Dakota": "Pierre",
    "Tennessee": "Nashville",
    "Texas": "Austin",
    "Utah": "Salt Lake City",
    "Vermont": "Montpelier",
    "Virginia": "Richmond",
    "Washington": "Olympia",
    "West Virginia": "Charleston",
    "Wisconsin": "Madison",
    "Wyoming": "Cheyenne"
}

states = list(us_states_capitals.keys())
capitals = list(us_states_capitals.values())

    
def rand_st_cap(us_states_capitals):
    ''' pick random state and its capital'''
    
    state, correct_capital = random.choice(list(us_states_capitals.items()))
    return (state,correct_capital)


def wrong_caps(us_states_capitals,correct_capital):
    '''removes correct capital from list, returns list of wrong capitals, and 
    picks 2 wrong capitals'''
    capitals = list(us_states_capitals.values())
    capitals.remove(correct_capital)#removes correct_capital from list 
    
    wrong_capitals = capitals 
    
    two_wrong_caps = random.sample(wrong_capitals,2)
    return two_wrong_caps

def shuffle_answers(two_wrong_caps,correct_capital):
    ''' shuffle all 3 capitals'''
    options = two_wrong_caps + [correct_capital]
    random.shuffle(options)
    return options

def answer_bank(options):
    '''display options as A, B, or C'''
    print('A = ' + options[0])
    print('B = ' + options[1])
    print('C = ' + options[2])
    



def question(state,options):
    '''quiz question. also gives player option to quit game before finishing'''
    acceptable_input = ['A','B','C','Q']
    answer = input(f"What is the capital of {state}?: Enter A, B, or C. If you want to quit game, enter Q: ").upper()
    
    if answer == 'Q':
        return 'Q'
    while answer not in acceptable_input:
        answer = input(f"Enter A, B, or C. If you want to quit game, enter Q: ").upper()
    return answer


def check_answer(answer,options,correct_capital):
    '''check if answer is correct'''
    letter_to_index = {'A':0,'B':1,'C':2}#used to give answer an index
    selected_option = options[letter_to_index[answer]]#inexing using words
    #finds answer index then uses it
    #to the corresponding element(option) at the same index in options list
   
    if selected_option == correct_capital:
        result = True
    else:
        result = False
    
    return result    




def final_score(correct,wrong):
    '''final score report'''
    q_total = correct + wrong #will allow to work if user quits game early
    score = float(correct/q_total)
    
    print(f'Your score is {score:.2%}')
    print(f'Correct answers: {correct}')
    print(f'Wrong answers: {wrong}')
    

def replay():
    '''allows player to replay'''
    ask = input('Play again? Y or N').upper()
    
    while ask not in ['Y','N']:
        print('Please enter Y or N')
        ask = input('Play again? Y or N')
    
    if ask == 'Y':
        states_and_caps_quiz(us_states_capitals)
    
    else:
        print('Bye')
```

```python
#runs game until each state has been asked
#game
def states_and_caps_quiz(us_states_capitals):
    '''runs quiz'''
    correct = 0 
    wrong = 0
    used_states = []
    
    print('This is a quiz that will test your knowledge of the US states and their capitals.')
    print('The answer bank for each question will appear above it.')
    print('Why? That is a very good question.')
    while len(used_states) != 50:

        (state,correct_capital) = rand_st_cap(us_states_capitals)
        while state in used_states: #makes sure each state is only used once
            (state,correct_capital) = random.choice(list(us_states_capitals.items()))
        
        else:
            used_states.append(state)
        
        two_wrong_caps = wrong_caps(us_states_capitals,correct_capital)
        
        options = shuffle_answers(two_wrong_caps,correct_capital)
        
        answer_bank(options)
        
        answer = question(state,options)
        
        if answer == 'Q':
            break
        

        result = check_answer(answer,options,correct_capital)
        if result == True:
            correct += 1 
        else:
            wrong += 1
    
    
    states_and_caps_quiz(us_states_capitals)
    
    final_score(correct, wrong)
    
    replay()
```
